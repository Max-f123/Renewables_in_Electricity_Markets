```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#set settings for plots
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 500
```

```python
one_price_results = pd.read_csv('1_3_results/one_price_results.csv')
one_price_results
```

```python
fig, ax = plt.subplots()
ax.scatter(one_price_results['betas'], one_price_results['objective_values'], label='objective_values')
ax.scatter(one_price_results['betas'], one_price_results['exp_profits'], label='exp_profits')
ax.set_xlabel('betas')
ax.set_ylabel('objective_values/exp_profits')
plt.legend( loc='upper left')


ax2 = ax.twinx()
ax2.scatter(one_price_results['betas'], one_price_results['CVaR_values'], label='CVaR_values', color='r')
ax2.set_ylabel('CVaR_values')

plt.legend(loc='center right')


plt.show()
```
