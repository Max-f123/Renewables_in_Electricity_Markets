```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

## One price scheme

```python
# load the data
data = pd.read_csv('1_5_results/relative_profits_one_price.csv')
```

```python
in_sample_scenarios = data['In sample scenarios']
seed = data['Seed']
in_sample_profit = data['In sample profit']
out_of_sample_profit = data['Out of sample profit']
relative_profit = data['Relative profit']
```

```python
# get the unique in sample scenarios
unique_in_sample_scenarios = data['In sample scenarios'].unique()

# get the mean of relative_profit for each in_sample_scenarios
mean_relative_profit = data.groupby('In sample scenarios')['Relative profit'].mean()

# get the standard deviation of relative_profit for each in_sample_scenario
std_relative_profit = data.groupby('In sample scenarios')['Relative profit'].std()
```

```python
# plot the same than above, but with a line for the mean, and an area for the standard deviation
plt.figure()
plt.plot(unique_in_sample_scenarios, mean_relative_profit, 'o')
plt.plot(unique_in_sample_scenarios, mean_relative_profit, color='b')
plt.axhline(y=0, color='r', linestyle='--')
plt.fill_between(unique_in_sample_scenarios, mean_relative_profit - std_relative_profit, mean_relative_profit + std_relative_profit, alpha=0.5)
plt.xlabel('Number of in sample scenarios')
plt.ylabel('Relative deviation between in and out of sample profit')
plt.title('Relative profit for 1-price')
plt.legend(['Standard deviation', 'Mean', 'Zero line'])
#plt.savefig('1_5_results/relative_profits_area.png')
plt.show()
```
