```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load files from result
reserve_market = pd.read_csv('results/reserve_market_clearing_model6.csv')
da_df = pd.read_csv('results/market_clearing_model6.csv')
da_df
```

```python
#plot the price columns as step function and the quanity columns as bar chart
fig, ax = plt.subplots()

plt.xticks(np.arange(1, 25, 3))

ax.step(reserve_market['hour'], reserve_market['market_price_up_reserve'], where='mid', label='Up Reserve Price', color='red')
ax.step(reserve_market['hour'], reserve_market['market_price_down_reserve'], where='mid', label='Down Reserve Price', color='blue')

ax2 = ax.twinx()
ax2.bar(reserve_market['hour'], reserve_market['reserve_req_up'], align = 'edge', width=-0.25, alpha=0.5, label='Up Reserve Quantity', color='orange')
ax2.bar(reserve_market['hour'], reserve_market['reserve_req_down'], align = 'edge', width=0.25, alpha=0.5, label='Down Reserve Quantity', color='blue')

ax.set_xlabel('Hour')
ax.set_ylabel('Price ($/MWh)')

ax2.set_ylabel('Quantity (MW)')

ax.legend(loc='center left', bbox_to_anchor=(1.1, 0.425))
ax2.legend(loc='center left', bbox_to_anchor=(1.1, 0.575))
plt.show()

#save plot
fig.savefig('results/reserve_market.png', bbox_inches='tight')
```

```python
# load equilibrium results from folder task 2
model2 = pd.read_csv('../Task2/results/equilibrium_model2.csv')
model2
```
