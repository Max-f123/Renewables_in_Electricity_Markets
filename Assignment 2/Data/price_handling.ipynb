```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = pd.read_excel('prices_excel.xlsx')
```

```python
# drop the first rows
prices = prices.drop([0, 1, 2, 3, 4, 5, 6, 7])
#use the first row as header
prices.columns = prices.iloc[0]
prices = prices.drop([8])
prices.reset_index(drop=True, inplace=True)
prices
```

```python
#change Datum to datetime format
prices['Date'] = pd.to_datetime(prices['Datum'], format='%d.%m.%Y')
#drop the hour part of date column
prices['Date'] = prices['Date'].dt.date

#get the first half of the string in Anfang column
prices['hour'] = prices['Anfang'].str.split(':').str[0].astype(int)
#drop Datum, Anfang and ende columns
prices = prices.drop(['Datum', 'Anfang', 'Ende'], axis=1)

prices.columns

#drop all prices except Dänemark 1
prices = prices.drop(['Deutschland/Luxemburg [€/MWh]', '∅ Anrainer DE/LU [€/MWh]',
       'Belgien [€/MWh]', 'Dänemark 2 [€/MWh]',
       'Frankreich [€/MWh]', 'Niederlande [€/MWh]', 'Norwegen 2 [€/MWh]',
       'Österreich [€/MWh]', 'Polen [€/MWh]', 'Schweden 4 [€/MWh]',
       'Schweiz [€/MWh]', 'Tschechien [€/MWh]', 'DE/AT/LU [€/MWh]',
       'Italien (Nord) [€/MWh]', 'Slowenien [€/MWh]', 'Ungarn [€/MWh]',], axis=1)
```
