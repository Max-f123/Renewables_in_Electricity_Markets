```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import result files
bat_df = pd.read_csv('results/bat_result.csv')
result_df = pd.read_csv('results/market_clearing_model2.csv')
equilbrium_df = pd.read_csv('results/equilibrium_model2.csv')
```

```python
equilbrium_df
```

```python
bat_df
#sum all columns which are containing char and sum all columns which are containing disch to get total charging and discharging
bat_df['total_charging'] = bat_df.filter(like='char').sum(axis=1)
bat_df['total_discharging'] = bat_df.filter(like='disch').sum(axis=1)
```

```python
# Create figure and primary axis
fig, ax1 = plt.subplots()

#x ticks should go from 1 to 24
plt.xticks(np.arange(0, 25, 3))

# Plot data on primary axis
ax1.step(equilbrium_df['hour'], equilbrium_df['market_price'], where='mid')
ax1.set_xlabel('Hour')
ax1.set_ylabel('Market price')  # Y-axis label for the primary axis

# Create and plot on secondary axis
ax2 = ax1.twinx()  # Create a second y-axis
ax2.bar(bat_df['hour'], bat_df['total_charging'], color='red', alpha=0.5)
ax2.bar(bat_df['hour'], bat_df['total_discharging'], color='green', alpha=0.5)
ax2.set_ylabel('MW battery charging/discharging')  # Y-axis label for the secondary axis

# Add a legend
ax1.legend(['Market price'], loc='center right')
ax2.legend(['Charging', 'Discharging'], loc='center right', bbox_to_anchor=(1, 0.6))

# Show plot
plt.show()

#save plot
fig.savefig('results/battery_market_clearing.png')
```
