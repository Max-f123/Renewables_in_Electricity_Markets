```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#set settings for plots
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 500
```

```python
# load the data
result = pd.read_csv('result_df.csv')
result['p_value'].values
result['shortfall/ bid ratio'] = result['below_also_x']/result['also_x_bids']
```

```python
#plot the results: p_value as x axis, also x bids as line chart and below also x as bar chart
fig, ax = plt.subplots()
ax.plot(result['p_value'].values, result['also_x_bids'], label='x bids', color='blue')
ax.set_xlabel('p_value')
ax.set_ylabel('also_x bids')
ax.set_xlim(result['p_value'].values.min()-0.01, result['p_value'].values.max()+0.01)
ax.set_xticks(result['p_value'].values)

ax.legend(loc='upper right')
ax2 = ax.twinx()
#ax2.plot(result['p_value'].values, result['shortfall/ bid ratio'], label='Shortfall/ bid ratio', color='green')
ax2.bar(result['p_value'].values, result['below_also_x'], width=0.01, label='Expected reserve shortfall', color='red')
ax2.set_ylabel('Expected reserve shortfall')
ax2.legend(loc='upper right', bbox_to_anchor=(1, 0.9))
plt.show()
```
