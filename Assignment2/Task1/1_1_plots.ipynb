```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#set settings for plots
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 500
```

```python
profits = pd.read_csv('1_1_results/profits_1_1.csv')

#make a latex table with the profits.describe() values
print(profits['profit'].describe().to_latex())
```

```python
#make good looking histogram of profits
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.hist(profits['profit'], bins=20)
plt.savefig('1_1_results/profits_histogram.png')
```
